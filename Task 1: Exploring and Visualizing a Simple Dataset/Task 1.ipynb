# 🌸 Floral Pattern Report: An Artistic Dive into the Iris Dataset
Ever wondered what secrets lie in the petals of flowers? Let's explore the famous Iris dataset — not as a data scientist — but as a curious botanist




**Import Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

**Load Dataset**

```python
df = sns.load_dataset("iris")
df.head()
```

**Data Structure**

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.describe()
```

**Data Visualization**

```python
#  Dark Theme Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Create custom dark theme
plt.style.use('dark_background')
sns.set_palette("rocket")
sns.set_context("notebook", font_scale=1.2)

# Optional: Customize figure background
plt.rcParams['axes.facecolor'] = '#111111'
plt.rcParams['figure.facecolor'] = '#111111'
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", s=100)
plt.title("🌌 Petal Geometry in the Dark: Stargazing the Species")
plt.show()
```

```python
# SCATTER PLOT: Sepal
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", palette="Set2")
plt.title("🧭 Sepal Dimensions: The Geometry of Growth")
plt.show()
```

```python
# HISTOGRAM: Petal Length
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="petal_length", hue="species", kde=True, palette="flare")
plt.title("🌿 Petal Length Distribution across Species")
plt.show()
```

```python
# BOX PLOT: Petal Width
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="species", y="petal_width", palette="pastel")
plt.title("🎁 Petal Width Spread: Unwrapping the Mystery")
plt.show()
```

```python
from wordcloud import WordCloud

# Simulate creative species names
tags = [
    "Setosaria", "VersaLegend", "Virginox", "SetoShine", "Versano", 
    "PetalPop", "Sepalix", "BlossomBound", "FloraZest", "BotanIQ"
]

# Join into single string for WordCloud
text = " ".join(tags * 10)  # Repeat to emphasize

# Generate WordCloud
wc = WordCloud(width=800, height=400, background_color="black", colormap="Pastel1")
plt.figure(figsize=(10,5))
plt.imshow(wc.generate(text), interpolation="bilinear")
plt.axis("off")
plt.title("🎨 Imaginative Iris Species Tag Cloud")
plt.show()
```
